```python
import matplotlib.pyplot as plt
        
def open_file(filename: str) -> list:
    with open(f"results/{filename}.txt", "r") as file:
        data = [float(line) for line in file]
        return data
        
    
```

```python
N = 100

indexs = [i for i in range(4, N + 1)]
result_classic = open_file("result/classic")
result_parallel = open_file("result/parallel")

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.plot(indexs, result_classic, color="blue", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_parallel, color="red", linewidth=1.0, linestyle="-")

ax.grid()
ax.set_xlabel(r'номер теста', fontsize=12)
ax.set_ylabel(r'сумма', fontsize=12)
ax.set_title("сравнение алгоритмов по результату")
ax.legend(["Classic", "Parallel"])
# plt.savefig("grafics/simulated and nearest both with population.png")
plt.show()
```

```python
N = 100
indexs = [i for i in range(4, N + 1)]
result_classic = open_file("time/classic")
result_parallel = open_file("time/parallel")

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.plot(indexs, result_classic, color="blue", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_parallel, color="red", linewidth=1.0, linestyle="-")

ax.grid()
ax.set_xlabel(r'номер теста', fontsize=12)
ax.set_ylabel(r'время (сек) ', fontsize=12)
ax.set_title("сравнение алгоритмов по времени работы")
ax.legend(["classic", "parellel"])
# plt.savefig("grafics/classic and modification time.png")
plt.show()
```

```python
# Данные для диаграммы
categories = ['Classic', 'Parallel']
values = [open_file("search/classic")[0], open_file("search/parallel")[0]]

# Цвета для каждого столбца
colors = ['blue', 'red']
# Создание столбчатой диаграммы
plt.figure(figsize=(15, 8))  # Установка размера фигуры
plt.bar(categories, values, color=colors)

# Добавление заголовка и подписей к осям
plt.title('Сравнение алгоритмов по вероятности нахождения гамильтонового цикла')
plt.xlabel('Алгоритмы')
plt.ylabel('Процент вероятности нахождения гамильтонового цикла')

# Отображение диаграммы
# plt.savefig("grafics/classic and my modification.png")
plt.show()

```
